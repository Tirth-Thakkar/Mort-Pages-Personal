```java
import java.util.Scanner;
import java.util.Random;

public class RockPaperScissorsGame {
    private static final String DEFAULT = "\u001B[0m";  // Default Terminal Color

    public static void main(String[] args) {
        playRPS();
    }

    public static void playRPS() {
        Scanner sc = new Scanner(System.in);
        
        System.out.println("Rock Paper Scissors");
        System.out.println("Type 'r' for rock, 'p' for paper, or 's' for scissors");
        
        Random random = new Random();
        char[] choices = {'r', 'p', 's'};
        char userChoice = sc.nextLine().toLowerCase().charAt(0);
        char computerChoice = choices[random.nextInt(3)];
        
        // Game logic
        if (userChoice == computerChoice) {
            System.out.println("It's a tie!");
        } else if (
            (userChoice == 'r' && computerChoice == 's') ||
            (userChoice == 's' && computerChoice == 'p') ||
            (userChoice == 'p' && computerChoice == 'r')
        ) {
            System.out.println("You win!");
        } else {
            System.out.println("Computer wins!");
        }
        
        sc.close();
    }
}
```

```java
import java.util.Scanner;

public class TicTacToeGame {
    public static void main(String[] args) {
        playTicTacToe();
    }

    public static void playTicTacToe() {
        Scanner sc = new Scanner(System.in);
        char[] board = {'1', '2', '3', '4', '5', '6', '7', '8', '9'};
        char currentPlayer = 'X';
        int moves = 0;
        boolean gameWon = false;

        System.out.println("Tic Tac Toe");
        System.out.println("Type the number of the square you want to place your piece in");
        displayBoard(board);

        while (moves < 9 && !gameWon) {
            System.out.println("Player " + currentPlayer + "'s turn");
            int move = sc.nextInt();

            if (move < 1 || move > 9 || board[move - 1] == 'X' || board[move - 1] == 'O') {
                System.out.println("Invalid move. Try again.");
                continue;
            }

            board[move - 1] = currentPlayer;
            moves++;
            displayBoard(board);

            // Check win conditions
            if (checkWin(board, currentPlayer)) {
                System.out.println("Player " + currentPlayer + " wins!");
                gameWon = true;
            }

            // Switch players
            currentPlayer = (currentPlayer == 'X') ? 'O' : 'X';
        }

        if (!gameWon) {
            System.out.println("It's a tie!");
        }

        sc.close();
    }

    private static void displayBoard(char[] board) {
        System.out.println(board[0] + " | " + board[1] + " | " + board[2]);
        System.out.println("---------");
        System.out.println(board[3] + " | " + board[4] + " | " + board[5]);
        System.out.println("---------");
        System.out.println(board[6] + " | " + board[7] + " | " + board[8]);
    }

    private static boolean checkWin(char[] board, char player) {
        // Check rows, columns, and diagonals for a win
        return ((board[0] == player && board[1] == player && board[2] == player) ||
                (board[3] == player && board[4] == player && board[5] == player) ||
                (board[6] == player && board[7] == player && board[8] == player) ||
                (board[0] == player && board[3] == player && board[6] == player) ||
                (board[1] == player && board[4] == player && board[7] == player) ||
                (board[2] == player && board[5] == player && board[8] == player) ||
                (board[0] == player && board[4] == player && board[8] == player) ||
                (board[2] == player && board[4] == player && board[6] == player));
    }
}
```

```java
import java.util.Scanner;
import java.util.Random;

public class HigherLowerGame {
    public static void main(String[] args) {
        playHigherLower();
    }

    public static void playHigherLower() {
        Scanner sc = new Scanner(System.in);
        Random random = new Random();
        int targetNumber = random.nextInt(8) + 1;
        int remainingAttempts = 3;

        System.out.println("Higher or Lower");
        System.out.println("You have three guesses to guess the number I'm thinking of between 1-8.");

        while (remainingAttempts > 0) {
            System.out.print("Enter your guess: ");
            int userGuess = sc.nextInt();

            if (userGuess == targetNumber) {
                System.out.println("You win!");
                break;
            } else if (userGuess < targetNumber) {
                System.out.println("The number is higher.");
            } else {
                System.out.println("The number is lower.");
            }

            remainingAttempts--;
        }

        if (remainingAttempts == 0) {
            System.out.println("Game over. The correct number was " + targetNumber);
        }

        sc.close();
    }
}
```

```java
import java.util.Scanner;

public class MainMenu {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        boolean quit = false;

        while (!quit) {
            displayMenu();
            int choice = sc.nextInt();

            switch (choice) {
                case 0:
                    System.out.println("Goodbye, World!");
                    quit = true;
                    break;
                case 1:
                    RockPaperScissors rps = new RockPaperScissors();
                    rps.play();
                    break;
                case 2:
                    HigherLower hl = new HigherLower();
                    hl.play();
                    break;
                case 3:
                    TicTacToe ttt = new TicTacToe();
                    ttt.play();
                    break;
                default:
                    System.out.println("Unexpected choice, try again.");
            }
        }

        sc.close();
    }

    public static void displayMenu() {
        String menuText =
            "___________________________\n" +
            "|~~~~~~~~~~~~~~~~~~~~~~~~~|\n" +
            "|          Menu!          |\n" +
            "|~~~~~~~~~~~~~~~~~~~~~~~~~|\n" +
            "| 0 - Exit                |\n" +
            "| 1 - Rock Paper Scissors |\n" +
            "| 2 - Higher or Lower     |\n" +
            "| 3 - Tic Tac Toe         |\n" +
            "|_________________________|\n\n" +
            "Choose an option: ";

        System.out.print(menuText);
    }
}
```
